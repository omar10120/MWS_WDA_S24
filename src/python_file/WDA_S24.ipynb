```python
# Delete unnecessary Columns
for column_name in ['ORDERDATE', 'ORDERLINENUMBER', 'CUSTOMERNAME', 'QTR_ID', 'PRODUCTLINE', 'MSRP',
'PHONE', 'ADDRESSLINE1', 'POSTALCODE', 'ADDRESSLINE2', 'COUNTRY', 'TERRITORY', 'STATE',
'CONTACTFIRSTNAME', 'CONTACTLASTNAME', 'DEALSIZE', 'STATUS']:
  DataSet = DataSet.drop(column_name, axis=1)
```

```python
#display information about the data
DataSet.info()
DataSet.describe(include='all')
```

```python
DataSet = DataSet[DataSet['YEAR_ID'] == DataSet['YEAR_ID'].max()]
```

```python
DataSet['SALES'] = DataSet['QUANTITYORDERED'] * DataSet['PRICEEACH']
```

```python
# check for duplicated values
DataSet.duplicated().sum()
```

```python
#display information about the data
DataSet.info()
DataSet.describe(include='all')
```

```python
#groups sales data by product code and calculates key metrics for each product
metrics_Products = DataSet.groupby('PRODUCTCODE').agg({'SALES': 'sum', 'QUANTITYORDERED': 'sum', 'PRICEEACH': 'mean', 'ORDERNUMBER': 'nunique'})
metrics_Products.rename(columns={'SALES': 'TOTAL_REVENUE', 'QUANTITYORDERED': 'TOTAL_UNITS_SOLD', 'PRICEEACH': 'AVG_PRICE_PER_UNIT', 'ORDERNUMBER': 'TOTAL_ORDERS'}, inplace=True)
```

```python
metrics_Products.describe()
```

```python
metrics_Products.sort_values(by='TOTAL_REVENUE', ascending=False, inplace=True)
print(metrics_Products.to_string())
```

```python
#calculates monthly sales metrics by grouping data by YEAR_ID and MONTH_ID.

monthly_metrics = DataSet.groupby(['YEAR_ID', 'MONTH_ID']).agg({'SALES': 'sum', 'QUANTITYORDERED': 'sum', 'PRICEEACH': 'mean'})
monthly_metrics.rename(columns={'SALES': 'TOTAL_REVENUE', 'QUANTITYORDERED': 'TOTAL_UNITS_SOLD', 'PRICEEACH': 'AVG_PRICE_PER_UNIT'}, inplace=True)
```

```python
#printing monthly_metrics
print(monthly_metrics)
```

```python
#summary of descriptive
monthly_metrics.describe()
```

```python
#generates a line plot of the TOTAL_REVENUE column from the monthly_metrics DataFrame,
#showing how total revenue changes over time
figure, axes = plt.subplots(figsize=(12,6))
monthly_metrics['TOTAL_REVENUE'].plot(label='Total Revenue')
axes.set_xlabel('Date')
axes.set_ylabel('Total revenue')
axes.set_title('Total revenue over time')
axes.legend()
y_min, y_max = plt.ylim()
plt.show()
```

```python
#generates a line plot of the TOTAL_UNITS_SOLD column from the monthly_metrics DataFrame,
#showing how the total number of units sold changes over time
import matplotlib.pyplot as plt
figure, axes = plt.subplots(figsize=(12,6))

monthly_metrics['TOTAL_UNITS_SOLD'].plot(label='Total Units Sold')

axes.set_xlabel('Date')
axes.set_ylabel('Total units sold')
axes.set_title('Total units sold over time')
axes.legend()
y_min, y_max = plt.ylim()
plt.show()
```

```python
#generates a line plot of the AVG_PRICE_PER_UNIT column from the monthly_metrics DataFrame,
#showing how the average price per unit changes over time.
import matplotlib.pyplot as plt
figure, axes = plt.subplots(figsize=(12,6))
monthly_metrics['AVG_PRICE_PER_UNIT'].plot(label='Average Price per Unit')
axes.set_xlabel('Date')
axes.set_ylabel('Average price per unit')
axes.set_title('Average price per unit over time')
axes.legend()
y_min, y_max = plt.ylim()
plt.show()
axes.grid(True, linestyle='--', alpha=0.7)
axes.set_ylim(y_min * 0.9, y_max * 1.1)  # Adds 10% padding above and below
monthly_metrics['AVG_PRICE_PER_UNIT'].rolling(window=3).mean().plot(label='3-Month Moving Avg', linestyle='--', ax=axes)
monthly_metrics['AVG_PRICE_PER_UNIT'].plot(label='Average Price per Unit', marker='o', ax=axes)
```
